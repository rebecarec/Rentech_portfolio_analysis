```python
import pandas as pd
import warnings
import yfinance as yf
from google.colab import files

warnings.filterwarnings('ignore') ### to kill the initial unecesary warnings

### Defining the tickersss
tickers = ['UTHR', 'PLTR', 'AAPL', 'KGC', 'MU', 'VRSN', 'EXEL', 'FNV', 'SNDK', 'NVDA']

### Downloading the tickers period
RT_current = yf.download( tickers=tickers,
                                start="2025-06-15",
                                end="2026-06-15",  # 12 months period analysis
                                interval="1d")

##########################################################################

# Create a dictionary to store metadata for each ticker
ticker_metadata = []

for ticker in tickers:
    try:
        RT_current_info = yf.Ticker(ticker).info
        # Extract desired fields with fallbacks in case data is missing
        sector = RT_current_info.get('sector', 'N/A')
        industry = RT_current_info.get('industry', 'N/A')

        ticker_metadata.append({
            'ticker': ticker,
            'sector': sector,
            'industry': industry
        })
    except Exception as e:
        print(f"Could not fetch data for {ticker}: {e}")
        ticker_metadata.append({'ticker': ticker, 'sector': 'N/A', 'industry': 'N/A'})

# Convert metadata list into a DataFrame
metadata_df = pd.DataFrame(ticker_metadata)

############################################### CREATING THE TABLE

print("--- RENTEC HISTORICAL CLOSE PRICES (REFORMATTED) ---")

##### Defining values for the new format
RT_current_close_prices = RT_current['Close']
RT_current_close_prices = RT_current_close_prices.reset_index()

RT_current_formatted_df = RT_current_close_prices.melt( id_vars=['Date'],
                                                        var_name='ticker',
                                                        value_name='close_value')    # The prices of the stocks per day
RT_current_formatted_df['ID'] = RT_current_formatted_df['Date'].astype(str) + "_" + RT_current_formatted_df['ticker']
RT_current_formatted_df = pd.merge(RT_current_formatted_df, metadata_df, on='ticker', how='left')

##### rearranging columns
RT_current_formatted_df = RT_current_formatted_df[['ID', 'ticker', 'Date', 'close_value','sector','industry']]

##### Print the final result
print(RT_current_formatted_df)

########################################## DONWLOAD WINDOW TO THE EXTRACTED FILE
filename = 'RT_current.csv'

#Save the file temporarily onto the Colab cloud server
RT_current_formatted_df.to_csv(filename, index=False)
#Trigger the browser to download it to your local machine
files.download(filename)
```
